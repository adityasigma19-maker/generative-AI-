```python
# variable length argument -- we give any number of argument
# non keyword variable length -- it is behave like tuple --- represent by star (*)
# keyword variable length argument -- it si behave like dictionary ---- repreeesent by double star (**)
# non keyword variable argument
def test(*args):
    print(args)
    print(len(args))
    print(type(args))
    test(1, 2, 3, 4, 5, 6)
```

```python
def sum_number(*args):
    sum = 0
    for num in args:
              sum+=num
    return sum
```

```python
# keyword variable length argument
def test(**kwargs):
  print(kwargs)
  print(len(kwargs))
  print(type(kwargs))
  test(a = 10, b = 20, c = 40)
```

```python
import numpy as np
from numpy.linalg import norm

A = np.array([2 ,3, 4, 5, 6 ,7])
B = np.array([8, 9, 1, 3, 5,9 ])

#compute cosin similarity
cosin = np.dot(A ,B) / (norm(A) * norm(B))
print("Cosine similarity :", cosin)
```

```python
from transformers import pipeline
summarizer = pipeline("summarization", model = "facebook/bart-large-cnn")
text = """the global shift toward renewable energy is accelerating ,
 driven by concern over climate change and energy security .
  Solar and wind power installation have reached record levels worldwide."""

summary = summarizer(text, max_length = 30, min_length = 10, do_sample = False)
print(summary[0])
```

```python
from transformers import AutoTokenizer
# Load a tokenizer (e.g , BERT )
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
text = "Machine learning is transforming material science!"
#Tokenize
tokens = tokenizer.tokenize(text)
token_ids = tokenizer.encode(text)

print("Token :", tokens)
print("Token IDs :", token_ids)
```
